```python
import pandas as pd
```

```python
dataset = pd.read_csv('gg_spending.csv')
```

```python
dataset.head()
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])

dataset['THB'].astype(int)
```

```python
dataset['day'] = dataset['Date'].apply(lambda x: x.day)
```

```python
dataset.tail()
```

```python
dataset = dataset.dropna()
```

```python
dataset['month'] = dataset['Date'].apply(lambda x: x.month)
```

```python
dataset['weekday_number'] = dataset['Date'].apply(lambda x: x.weekday() + 1)
```

```python
dataset
```

```python
dataset['quarter'] = dataset['Date'].apply(lambda x: x.quarter)
```

```python
dataset['weekday'] = dataset['weekday_number'].map({1:'Monday',
                                   2:'Tuesday', 
                                   3:'Wednesday', 
                                   4:'Thursday',
                                   5:'Friday',
                                   6:'Satrday',
                                   7:'Sunday'})
```

```python
dataset.pivot_table(index='weekday',columns='month', values='THB', aggfunc='mean')
```
